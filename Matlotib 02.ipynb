```python
plt.plot([1, 2,  3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
Parameter ketiga pengatur bentuk marker yang dimaksud adalah karakter o dan bukan nol
```

```python
plt.plot([1, 2,  3, 4], [1, 4, 9, 16], 'g*')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
Multi plot dalam satu pemanggialn fungsi plot()
```

```python
t = np.arange(0., 5., 0.2)
t
```

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
plt.plot(t, t, 'r--',
        t, t**2, 'bs',
        t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
       'c': np.random.randint(0, 50, 50),
       'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entery a')
plt.ylabel('entery b')
plt.show()
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entery a')
plt.ylabel('entery b')
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
Plotting untuk tipe data categorial

Matplotlib mendukung plotting untuk data dengan tipe kategori atau biasa dikenal dengan istilah categorical data categorical variables.
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
```

```python
plt.plot(names, values)
plt.show()
```

```python
plt.scatter(names, values)
plt.show()
```

```python
plt.bar(names, values)
plt.show()
```
